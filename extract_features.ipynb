```python
import pandas as pd
from gazpacho import Soup, get
import ast
# import urllib.request
import requests
```

```python
df = pd.read_csv('country_with_url_pages_notes.csv')
df.head()
```

```python
for i in ast.literal_eval(df.NOTES[0]):
    print(i)
```

```python
# count = 0
# features_country = []
# photos_availability = []

notes_each_country = []
for row in df.iterrows():
    notes = ast.literal_eval(row[1]['NOTES'])
    country = row[1]['Country_Name']
    feature_each_note = []
    photos_each_note = []
    print("COUNTRY:", country)
    for note in notes:
        try:
            http = Soup.get(note)
        except:
            print(note)
            break
        div_photos = http.find('div', attrs={"id":"fiche_photo"})
        photos = div_photos.find('a')
        ref = note[37:-5]
        # headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'}
        # print("REFERENCE:",ref)

        features = http.find('section', attrs={'id':'fiche_caracteristiques'}).find("tr")
        feature_table = {}
        feature_table["Country"] = country
        feature_table["Reference"] = ref
        for feature in features:
            heading = feature.find('th').text
            if type(feature.find('td').find('a')) != list:
                value = feature.find('td').text
            else:
                value = feature.find('td').text
                for idx, val in enumerate(feature.find('td').find('a')):
                    if idx > 0:
                        value += " ("+val.text +")"
            feature_table[heading] = value
        # print("DONE:",ref, features_country)
        # feature_each_note.append(features_country)

        photo_list = []
        if photos:
            if type(photos) != list:
                photos = [photos]
            # photos_each_note.append(True)
            for index, photo in enumerate(photos):
                photo_soup = photo.attrs['href']
                if photo_soup.endswith('.jpg') or photo_soup.endswith('.jpeg') or photo_soup.endswith('.png'):
                    photo_list.append(photo_soup)
                # if photo_soup.endswith('.jpg'):
                #     new_name = "./currencies/country_"+ref+"_"+str(index)+".jpg"
                #     print(country,':',photo_soup, ':', new_name)
                #     # urllib.request.urlretrieve(photo_soup, new_name)
                #     res = requests.get(photo_soup,headers=headers)
                #     with open(new_name, 'wb') as f:
                #         f.write(res.content)
                # elif photo_soup.endswith('.jpeg'):
                #     new_name = "./currencies/country_"+ref+"_"+str(index)+".jpeg"
                #     print(country,':',photo_soup, ':', new_name)
                #     res = requests.get(photo_soup,headers=headers)
                #     with open(new_name, 'wb') as f:
                #         f.write(res.content)
                # elif photo_soup.endswith('.png'):
                #     new_name = "./currencies/country_"+ref+"_"+str(index)+".png"
                #     print(country,':',photo_soup, ':', new_name)
                #     res = requests.get(photo_soup,headers=headers)
                #     with open(new_name, 'wb') as f:
                #         f.write(res.content)
        else:
            photos_each_note.append(False)
        feature_table["Photos"] = photo_list
        notes_each_country.append(feature_table)
    # features_country.append(feature_each_note)
    # photos_availability.append(photos_each_note)
            

    
    
```
