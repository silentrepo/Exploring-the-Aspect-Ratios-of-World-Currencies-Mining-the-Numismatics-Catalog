```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


df_filter['Length'] = pd.to_numeric(df_filter['Length'], errors='coerce')
df_filter['Width'] = pd.to_numeric(df_filter['Width'], errors='coerce')
df_filter.dropna(subset=['Length', 'Width'], inplace=True)


scaler = StandardScaler()
df_filter[['Length', 'Width']] = scaler.fit_transform(df_filter[['Length', 'Width']])


wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df_filter[['Length', 'Width']])
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
df_filter['Cluster'] = kmeans.fit_predict(df_filter[['Length', 'Width']])

plt.figure(figsize=(10, 8))
plt.scatter(df_filter[df_filter['Cluster'] == 0]['Length'], df_filter[df_filter['Cluster'] == 0]['Width'], s=100, c='red', label='Cluster 1')
plt.scatter(df_filter[df_filter['Cluster'] == 1]['Length'], df_filter[df_filter['Cluster'] == 1]['Width'], s=100, c='blue', label='Cluster 2')
if optimal_clusters > 2:
    plt.scatter(df_filter[df_filter['Cluster'] == 2]['Length'], df_filter[df_filter['Cluster'] == 2]['Width'], s=100, c='green', label='Cluster 3')

plt.title('Clusters of Currencies')
plt.xlabel('Length (Standardized)')
plt.ylabel('Width (Standardized)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
ratios = {
    "root-two": np.sqrt(2),
    "two root-two": 2 * np.sqrt(2),
    "two": 2,
    "three root-two": 3 * np.sqrt(2),
    "root-three-over-two": (np.sqrt(3) / 2),
    "root-three": np.sqrt(3),
    "two root-three": 2 * np.sqrt(3),
    "root-five-over-two": (np.sqrt(5) / 2),
    "root-five": np.sqrt(5),
    "golden ratio φ": (1 + np.sqrt(5)) / 2
}

def closest_ratio(aspect_ratio, ratio_dict):
    min_diff = float("inf")
    closest = None
    for name, value in ratio_dict.items():
        diff = abs(aspect_ratio - value)
        if diff < min_diff:
            min_diff = diff
            closest = name
    return closest

df_region['Closest Ratio'] = df_region['Aspect Ratio'].apply(lambda x: closest_ratio(x, ratios))

display(df_region.head())
```

```python
df_region['Year'] = pd.to_numeric(df_region['Year'], errors='coerce')
df_region['Year'] = pd.to_datetime(df_region['Year'], format='%Y')

# Assign closest ratio to each row
df_region['Closest Ratio'] = df_region['Aspect Ratio'].apply(lambda x: closest_ratio(x, ratios))

# Group data to track transitions
df_transition = df_region.groupby(['Year', 'Closest Ratio']).size().unstack().fillna(0)

# Plot global transition over time
plt.figure(figsize=(12, 6))
df_transition.plot(kind='line', marker='o', figsize=(12, 6), colormap='viridis')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Global Transition of Aspect Ratios Over Time')
plt.legend(title='Closest Ratio')
plt.grid(True)
plt.show()

# Transition per continent
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure Year is in datetime format
df_region['Year'] = pd.to_numeric(df_region['Year'], errors='coerce')
df_region['Year'] = pd.to_datetime(df_region['Year'], format='%Y')

# Aggregate: Compute average currency area by Continent & Year
df_time_series = df_region.groupby(['Year', 'Continent'])['Area'].mean().reset_index()

# Get unique continents
continents = df_time_series['Continent'].unique()

# Create subplots
fig, axes = plt.subplots(nrows=len(continents), ncols=1, figsize=(10, 5 * len(continents)), sharex=True)

# If only one subplot, wrap axes in a list for consistent iteration
if len(continents) == 1:
    axes = [axes]

# Plot each continent separately
for ax, continent in zip(axes, continents):
    subset = df_time_series[df_time_series['Continent'] == continent]
    ax.plot(subset['Year'], subset['Area'], marker='o', linestyle='-')
    ax.set_title(f'Average Currency Area in {continent}')
    ax.set_ylabel('Area (mm²)')
    ax.grid(True)

# Common x-axis label
axes[-1].set_xlabel('Year')

plt.tight_layout()
plt.show()
```
