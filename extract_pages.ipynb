```python
import pandas as pd
from gazpacho import Soup
```

```python
countries = pd.read_csv('country_with_url.csv')
countries
```

```python
pages = []

for row in countries['Country_Link']:
    country_link = Soup.get(row)
    navs = country_link.find('nav')
    if navs and navs[-1].text.__contains__('Pages'):
        pages.append(int(navs[-1].find('a')[-2].text))
    else:
        pages.append(1)
```

```python
countries['Pages'] = pages
```

```python
country_pages_list = []
country_name = []
for row in countries.iterrows():
    link = row[1]['Country_Link'][:-6]
    pages = row[1]['Pages']
    country_name.append([row[1]['Country_Name']][0])
    country_pages = []
    for i in range(1,pages+1):
        link_new = link + str(i) +'.html'
        country_pages.append(link_new)
    country_pages_list.append(country_pages)
```

```python
country_name,  country_pages_list
```

```python
df = pd.DataFrame(zip(country_name, country_pages_list), columns=['Country_Name', 'Country_Links'])
```
