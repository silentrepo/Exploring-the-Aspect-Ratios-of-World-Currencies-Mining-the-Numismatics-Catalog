```python
import pandas as pd
from gazpacho import Soup
```

```python
countries = pd.read_csv('country_with_url_pages_included.csv')
countries
```

```python
import ast
```

```python
countries.head()
```

```python
country_ref = []
for row in countries.iterrows():
    links = ast.literal_eval(row[1]['Country_Links'])
    country = row[1]['Country_Name']
    print(country)
    ref = []

    for link in links:
        page = Soup.get(link)
        divs = page.find('div', attrs={"class":"description_piece"})
        if type(divs) != list:
            ref.append(div.strip('br'))
        else:
            for div in divs:
                ref.append(div.strip('br'))
    country_ref.append(ref)
```

```python
countries["Country_Ref"] = country_ref
```
