```python
import pandas as pd
from gazpacho import Soup
import ast
```

```python
df = pd.read_csv('country_with_url_pages_refs.csv')
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
df
```

```python
country_ref_num = []
PREFIX = "https://en.numista.com/catalogue/note"
for row in df.iterrows():
    ref_nums = []
    refs = ast.literal_eval(row[1]['Country_Ref'])
    for ref in refs:

        ref_nums.append(PREFIX+ref.split(',')[-1].split('# ')[-1].strip()+'.html')
    country_ref_num.append(ref_nums)
    
```

```python
country_ref_num
```

```python
df['NOTES'] = country_ref_num
```

```python
df.to_csv('country_with_url_pages_notes.csv')
```
